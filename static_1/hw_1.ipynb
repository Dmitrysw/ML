### 1

Выберите все пары независимых событий:
- A = {число оканчивается на 6}
- B = {сумма цифр чётная}
- С = {число чётное}

Рассмотрим каждую пару событий и проверим независимость по определению: $P(A \mid B) = P(A)$.

Посчитаем априорные вероятности:
1. $P(A) = \frac{1}{10}$ — так как возможно 10 равновероятных вариантов последней цифры в числе.
1. $P(B) = \frac{1}{2}$ — так как возможно 2 равновероятных вариантов (сумма цифр либо чётная, либо нечётная).
1. $P(C) = \frac{1}{2}$ — так как возможно 2 равновероятных вариантов (число либо чётное, либо нечётное).

Посчитаем условные вероятности:
1. $P(A \mid B) = \frac{1}{10}$ — возможно всё так же 10 равновероятных вариантов последней цифры в числе, даже если сумма цифр чётная.
1. $P(A \mid C) = \frac{1}{5}$ — возможно уже не 10 вариантов последней цифры в числе, а только 5, так как событие C уже ограничивает нас множеством последних цифр (0, 2, 4, 6, 8).
1. $P(B \mid C) = \frac{1}{2}$ — возможно всё так же 2 равновероятных варианта.

Таким образом, получаем, что **независимые пары тут A и B, B и C**. А вот пара A и C независимой не является, так как не выполняется равенство.

Иногда бывает сложно сходу посчитать условную вероятность. В таких случаях можно пользоваться формулой условной вероятности: $P(A \mid B) \cdot P(B) = P(A \cap B)$.

### 2

На экзамене есть 10 билетов, из которых 2 простые для студентов. У студента есть возможность достать 2 билета (одновременно) и отвечать по одному из них (студенты предпочитают простые билеты). Какова вероятность, что отвечать придётся на простой билет?

Задачу можно решить двумя способами.

**Первый способ**

Нужная нам вероятность:

P(хотя бы один билет простой) = P(оба билета простые) + P(ровно один билет простой) — так как взаимоисключающие события и их пересечение имеет нулевую вероятность.

Посчитаем каждое слагаемое:

P(оба билеты простые) = P(первый билет простой) * P(второй билет простой | первый билет простой) = 2/10 * 1/9 = 1/45 — после того, как мы достанем простой билет, останется только 9 билетов и только 1 из них простой

P(ровно один билет простой) = P(первый билет простой, второй билет сложный) + P(первый билет сложный, второй билет простой) = 2/10 * 8/9 + 8/10 * 2/9 = 32/90 = 16/45 — опять же не забываем, что после извлечения одного билета сумма меняется

Сложим и **получим 17/45**.

**Второй способ**

P(хотя бы один билет простой) = 1 - P(оба билета непростые) = 1 - 8/10 * 7/9 = 1 - 56/90 = 34/90 = 17/45.

### 3

Иван ждёт друзей на празднование Дня рождения. Мария обещала приготовить торт, но с вероятностью 0.2 у неё может не получиться крем (в противном случае торт будет успешно приготовлен). Руслан собирался приготовить салат, но с вероятностью 0.4 он не найдёт на рынке свежих овощей. Иван знает, что ребята с вероятностью 0.5 приготовят успешно оба блюда.

С какой вероятностью на столе у Ивана окажется хотя бы одно блюдо?

**Решение**

Сначала воспользуемся свойством вероятности противоположного события, чтобы оценить по отдельности вероятности успешного приготовления салата и торта:

P(будет приготовлен торт) = 1 - P(крем не получится) = 1 - 0.2 = 0.8

P(будет приготовлен салат) = 1 - P(на рынке не будет овощей) = 1 - 0.4 = 0.6

Теперь воспользуемся свойством вероятности объединения двух событий:

P(будет приготовлено одно из блюд) = P(будет приготовлен салат) + P(будет приготовлен торт) - P(будет приготовлен салат и торт) = 0.8 + 0.6 - 0.5 = 0.9

### 4

Предположим, что у нас есть интернет-магазин и мы договорились поддержать ключевого партнёра. В обычной ситуации пользователи покупают товар этого партнёра с вероятностью 0.05 (без специальных показов). Если мы специально покажем товар на главной странице, то вероятность покупки составит 0.2.

Мы пообещали партнёру 1000 продаж товара за день, а трафик в день — 10000 пользователей (один пользователей делает ровно одну покупку в день). С какой минимальной вероятностью надо показывать товар на главной, чтобы продажи соответствовали нашим договорённостям?

**Решение**

Сначала определим нашу целевую вероятность покупки — 1000 продаж на 10000 посещений, то есть 0.1.

Распишем по формуле полной вероятности эту целевую вероятность покупки:

P(купит) = P(купит | показали на главной) P(показали на главной) + P(купит | не показали на главной) P(не показали на главной)

По свойству вероятности противоположных событий мы знаем, что:

P(не показали на главной) = 1 - P(показали на главной). 

В итоге получаем уравнение с одной неизвестной P(показали на главной), которую и хотим найти.

0.1 = 0.2 * x + 0.05 * (1-x) 
0.15x = 0.05
**x = 1/3**

### 8

Представим, что мы работаем с музыкальным сервис с рекомендациями. Для любой песни, попавшей в плейлист пользователей можно сказать, что половина пользователей послушает первые 20 секунд трека, из оставшейся половины две трети пользователей послушает трек ровно один раз, а одна треть — 3 раза. Длина трека — 4 минуты.

Посчитайте математическое ожидание продолжительности прослушивания трека на пользователя в секундах.

**Решение**

Посчитаем математическое ожидание по формуле:

$\mathbb{E}x$ = 20 * P(пользователь, который слушал 20 секунд) + 1 * 4 * 60 * P(пользователь прослушал ровно 1 раз) + 3 * 4 * 60 * P(пользователь прослушал ровно 3 раза) = 20 * 1/2 + 4 * 60 * (1/2 * 2/3) + 3 * 4 * 60 * (1/2 * 1/3) = 210

### 9

Предположим, что у нас есть интернет-магазин с товарами для квартир и частных домов. К нам приходят посетители двух типов — те, которые в поисках товаров для квартир и те, кто хотят купить товары для частных домов.

Априорная вероятность того, что новый посетитель ищет товары для квартиры — 0.8. Если посетитель ищет товары для квартиры, то первый клик он делает на товары для квартир с вероятностью 0.9. Если посетитель пришёл в поисков товаров для частного дома, то первый клик он делает на товары для частных домов с вероятностью 0.7.

Изначально мы ничего не знаем про нового посетителя и не можем построить какой-либо рекомендательной системы. Поэтому хотим по первому клику оценивать апостериорную вероятность посетителя, чтобы далее показывать подборку популярных товаров в таком соотношении вместо использования априорного распределения.

Посчитайте вероятность того, что посетитель пришёл за товарами для частного дома, если первый клик он сделал по товару для квартир (все варианты даны с округлением до второго знака после запятой).

**Решение**

Для начала определимся с тем, что нам надо посчитать. Апостериорная вероятность пользователя при условии клика: 

P(владелец дома | клик на квартирные товары) = 1 - P(владелец квартиры | клик на квартирные товары) — нам будет удобнее искать такую вероятность.

Для поиска этой вероятности нам понадобится теорема Байеса:

P(владелец квартиры | клик на квартирные товары) = P(клик на квартирные товары | владелец квартиры) P(владелец квартиры) / P(клик на квартирные товары) — здесь мы знаем два первых множителя из условия задачи и не знаем последней вероятности.

Для поиска верояности клика на квартирные товары воспользуемся формулой полной вероятности:

P(клик на квартирные товары) = P(клик на квартирные товары | владелец квартиры) P(владелец квартиры) + P(клик на квартирные товары | владелец дома) P(владелец дома).

Вероятность клика на квартирные товары у владельца дома мы не знаем, зато знаем вероятность противоположного события (клик на товары для дома при условии владельца дома).

Начнём собирать вcё вместе:
1. P(клик на квартирные товары) = 0.9 * 0.8 + (1 - 0.7) * 0.2 = 0.78
2. P(владелец квартиры | клик на квартирные товары) = 0.9 * 0.8 / 0.78 = 0.92
3. P(владелец дома | клик на квартирные товары) = 1 - 0.92 = 0.08

Проверим интуитивно: до того, как мы что-либо знали о посетителе, то вероятность того, что он пришёл за товарами для дома составляла 0.1. Как только он кликнул на товары для квартиры (что обычно делают владельцы квартир), вероятность уменьшилась до 0.08.
