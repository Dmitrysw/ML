```python
import numpy as np
import pandas as pd
import hashlib
```

```python
np.random.seed(6)

users_num = 10000
mean_user_ctr = 0.2
beta = 20
alpha = mean_user_ctr * beta / (1 - mean_user_ctr)

df = pd.DataFrame()
df['user'] = range(users_num)
df['group'] = np.random.rand(users_num) < 0.5

df['base_user_ctr'] = np.random.beta(alpha, beta, size=users_num)
df['views'] = np.random.lognormal(mean=1, sigma=1, size=users_num).astype(int) + 1
df['clicks'] = np.random.binomial(df['views'], df['base_user_ctr'])
```

```python
df.head(20)
```

```python
df["bucket"] = df['user'].apply(lambda x: int(hashlib.md5((str(x) + "my_salt").encode()).hexdigest(), 16) % 100)
```

```python
new_df = df.groupby(['bucket']).agg({'views': 'sum', "clicks" : "sum", 'base_user_ctr': 'mean'}).reset_index()
new_df['group_ctr'] = new_df.clicks / new_df.views
```

```python
std_base = new_df.loc[new_df['group'] == True, 'base_user_ctr'].std()
std_group = new_df.loc[new_df['group'] == True, 'group_ctr'].std()
```
