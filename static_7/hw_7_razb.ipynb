#### 1

Что из этого сложно или невозможно проверить через опросы пользователей?

- Какой функциональности не хватает в приложении
- [+] Качество подбора текста в пуш-сообщениях
- [+] Новую систему ранжирования поисковой выдачи
- [+] Новый алгоритм построения маршрутов навигатора

**Комментарий**

Хоть и кажется, что пользователя можно спросить про новое ранжирование поиска или алгоритм построения маршрута, но на самом деле вы не можете честно дать сравнить два алгоритма пользователям, чтобы они не просто выбрали наилучший, но и пережили одинаковый опыт — нельзя дать человеку проехать по двумя версиям маршрута в одинаковых условиях.

Пуш-уведомления нужно тоже проверять через A/B эксперимент.

А вот про недостающую функциональность пользователей можно спросить, так как сложно выкатить всё и проверить, что будет востребовано.

#### 2

В чём идея A/B тестирования?

- Дать пользователям сначала одну версию приложения, а потом другую — спросить пользователей, где было лучше
- Показать пользователю одновременно две версии маршрута в навигаторе — спросить, какая лучше
- [+] Дать разным группам пользователей разные версии приложения — сравнить метрики между группами

**Комментарий**

Суть A/B тестирования в том, чтобы взять две независимых группы пользователей, включить для каждой один из алгоритмов и сравнить метрики между собой. 

#### 3

Назовите неотъемлемые части A/B тестирования

- Чередование тестируемых алгоритмов на пользователях
- [+] Тестирование производится на двух или более независимых группах пользователей
- [+] Тестирование производится на репрезентативном периоде времени и достаточном количестве наблюдений
- [+] При сравнение метрик между группами нужно оценивать значимость отклонений

**Комментарий**

В A/B тестирование важно как раз запустить разные версии алгоритма или приложения на независимых группах пользователей, чтобы каждая из них использовала ровно одну версию алгоритма, поэтому чередование как раз не является верным ответом.

Так как это две разных группы пользователей, то метрики будут почти всегда различаться, надо оценивать значимость отклонений. 

Данных (наблюдений) должно быть достаточно и при этом данные должны быть репрезентативные (то есть быть похожи на генеральную совокупность).

#### 4

Если основные метрики в продукте чаще всего средний чек пользователя или конверсия в заказ, то зачем нужны вспомогательные метрики в A/B тестировании?

- [+] Вспомогательные метрики позволяют лучше понять, что происходит на разных этапах взаимодействия и какой эффект приносит изменение
- [+] Основные денежные метрики недостаточно чувствительны для небольших изменений продукта, вспомогательные — более чувствительны
- Вспомогательные метрики на самом деле не нужны — есть основные бизнес-метрики, их хватает
- Чем больше метрик, тем лучше — проще найти положительный эффект от внедрения

**Комментарий**

Денежные метрики имеют большую дисперсию и являются менее чувствительными. Также если мы делаем небольшие изменения, то вряд ли стоит ожидать сразу изменение среднего чего. В этом нам помогают вспомогательные метрики — они более чувствительны, так как описывают отдельные этапы взаимодействия в приложении и позволяют лучше понять, что происходит в эксперименте.

#### 5

Представим, что мы разрабатываем навигатор и алгоритмы для построения маршрутов в нём. Пользователи в приложении выбирают точку назначения из поиска или последних точек, строят маршрут и едут по нему. Если маршрут их не устроил, они могут выйти из приложения, найти другую точку и построить маршрут до неё, выбрать не основной, а альтернативный маршрут. Явной монетизации (подписки или рекламы) в приложении пока нет. Какая метрика наиболее подходит на роль основной для нашего продукта?

- Количество поисков на сессию
- Количество построений маршрутов на сессию
- Доля поездок, завершённых с отклонением от исходного маршрута
- [+] Конверсия из захода приложения в совершённую поездку по маршруту

**Комментарий**

Количество построений маршрутов на сессию — хорошая метрика, чтобы понимать, что маршруты строятся недостаточно хорошо (раз люди их перестраивают), но не показывает, как часто люди по ним потом уезжают (например, если первый раз построить маршрут плохо, то люди могут и не пытаться повторить или исправить).

Количество поисков на сессию — метрика, показывающая, как хорошо получаются найти точку назначения у пользователей, но никак не оценивает качество самих маршрутов.

Доля поездок, завершённых с отклонением от исходного маршрута — метрика, показывающая, насколько хорошие маршруты получаются (люди по ним проезжают), но упускает случай, где человек даже не начал поездку по неудачному маршруту.

Конверсия из захода приложения в совершённую поездку по маршруту — метрика, опивающая всё взаимодействие с приложением от захода до финального действия, как раз по ней можно оценивать все изменения, в том числе качество построения маршрутов.

#### 6

Продолжим развивать навигатор. Представим, что наша команда работает только над улучшением качества маршрутов. Какие вспомогательные метрики нам подойдут?

- [+] Конверсия из построения маршрута в поездку
- [+] Доля поездок, совершённых без отклонений от исходного маршрута
- [+] Количество поисков точки назначения на сессию
- [+] Доля сессий, где пользователь выбирал альтернативный маршрут

**Комментарий**

Количество поисков точки назначения на сессию — может показаться, что эта метрика не релевантна, однако если людям не нравится маршрут, то они могут пытаться выбрать другую точку назначения, чтобы попробовать получить другой маршрут.

Остальные метрики тоже показывают то, насколько хороший маршрут получился (люди не выбирают альтернативный, люди по ним ездят и не отклоняются от них). Каждая из них оценивает свою часть флоу пользователя в приложении.

#### 7

Представим, что вы улучшали приложение для заказа продуктов и проверяете, стали ли люди тратить меньше времени на совершение заказа, то есть измеряете продолжительность сессии пользователя в секундах.

Вы проводили A/B тестирования с двумя группами пользователей, одна из которых пользовалась обновлённой версией приложения. Вы собрали метрики по всем пользователям и знаете, кто из них в какую группу эксперимента попал.

Как вы оцените результаты?

- [+] Применим t-критерий Стьюдента для проверки гипотезы о равенстве средних продолжительность сессий — если гипотеза будет отвергнута, то сделаем вывод, что разница есть
- Посчитаем среднее значение метрик по пользователям в каждой из групп, сравним значения — где меньше, та группа и лучше
- Применим z-критерий для долей для проверки гипотезы о равенстве средних продолжительность сессий — если гипотеза будет отвергнута, то сделаем вывод, что разница есть
- [+] Построим доверительные интервалы на значения метрик в каждой из групп (или доверительный интервал на разность значений) — если интервалы пересекаются, то не можем сказать, что есть разница

**Комментарий**

При проведении A/B тестирования мы собираем метрики с разных пользователей, поэтому почти всегда они будут различаться. Нельзя сравнивать метрики напрямую и делать выводы о наличии эффекта.

Доверительные интервалы как и статистические критерии — способ оценивания достоверности изменений и отличия их от случайного отклонения. Однако z-критерий для долей — статистический критерий для бинарных метрик, здесь мы им не сможем воспользоваться. 

#### 8

Когда мы проводим A/B эксперименты и сравниваем метрики между группами, то пользуемся статистическими критериями. В таком случае у нас возможны ложноположительные (посчитали, что есть изменение, хотя его нет) и ложноотрицательные (не заметили изменения) ошибки.

Что мы делаем для того, чтобы ложноположительных ошибок (ошибок первого рода) было как можно меньше?

- Выбираем статистические критерии, которые минимизируют ошибку первого рода
- Берём больше наблюдений, так как ошибка первого рода зависит от числа наблюдений
- [+] Фиксируем ошибку первого рода на том значении, которое нас устроит, но сделать нулевой не можем
- Мы не можем повлиять на ошибку первого рода

**Комментарий**

Процедура оценки статистической значимости основана на том, что критические изменения в метрики при верности нулевой гипотезы (обычно, равенство средних) мы считаем за значимые отклонения. Таким образом, всегда существует ошибка первого рода, её уровень (уровень значимости, который мы сравниваем с p-value) мы выбираем сами (обычно это 0.05 или 0.01).

А теперь обратимся к нашему проекту, в котором вы разрабатываете рекомендательную систему для постов.

Какая метрика на ваш взгляд наиболее подходит на роль основной для этого продукта? Представим, что явного способа монетизации пока нет, чтобы посчитать денежную выручку.

- [+] Время, которое проводят пользователи в приложении (смотрят на рекомендации постов и читают сами посты)
- Количество кликов на пользователя по постам
- Количество постов, которые рекомендуются пользователям хотя бы раз

**Комментарий**

Метрики, связанные с кликами могут быть обманчивы — если ваша модель будет находить кликбейтные заголовки, то пользователи будут чаще на них кликать, но при этом не читать сами статьи (а ещё обычно долгосрочно люди перестают пользоваться рекомендациями, которые подсовывают им кликбейт). 

Метрика timespend (время, проведённое в приложении), как раз одновременно описывает и то, насколько люди залипают в рекомендации (им интересно находить в рекомендациях интересные посты) и сколько они читают эти посты при этом.

А теперь давайте подумаем про вспомогательные метрики для рекомендаций постов.

Какие из них подойдут, чтобы оценивать качество рекомендаций с разных сторон?

- [+] Количество постов, которые рекомендуются пользователям хотя бы раз
- [+] Доля сессий с кликами на один из первых трёх рекомендованных постов
- [+] Количество кликов по постам на пользователя

**Комментарий**

Все эти метрики релевантны для рекомендательных систем. В том числе метрика про разнообразие наших рекомендаций — важный показатель того, что мы не оказываемся в пузыре из показа одних и тех же постов.
