```python
import matplotlib.pyplot as plt
import math

# Создаем список значений для x и y
x = range(1, 120)
y = [math.factorial(i) for i in x]

# Строим график
plt.plot(x, y)

# Добавляем заголовок и метки осей
plt.title('График функции x!')
plt.xlabel('x')
plt.ylabel('y')

# Отображаем график
plt.show()
```

```python
#O(n)
def loop(n):
    res = 0
    for i in range(n):
        res += i
    return res
```

```python
def simplest(n):
    #O(1)
    return n << 2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаем массив значений x от 0 до 10 с шагом 0.1
x = np.arange(0, 1000, 0.1)

# Вычисляем значения функций y = sqrt(x) и y = ln(x)
y1 = x ** (1/3)
y2 = np.log(x)

# Создаем график
plt.plot(x, y1, label='sqrt(x)')
plt.plot(x, y2, label='ln(x)')

# Добавляем заголовок графика и подписи осей
plt.title('График функций y = sqrt(x) и y = ln(x)')
plt.xlabel('x')
plt.ylabel('y')

# Добавляем легенду
plt.legend()

# Отображаем график
plt.show()
```

```python
def gen_pairs(arr):
    result = []
    for i, elem in enumerate(arr):
        for elem_2 in arr[i:]:
            result.append((elem, elem_2))
    return result
```
